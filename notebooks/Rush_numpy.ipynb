```python
#Filter_RISE
def Euclidist(mat, m, n):
    center_x = m // 2
    center_y = n // 2
    i = 0
    for i in range(0, n):
        for j in range(0, m):
            mat[j, i] = np.sqrt((i - center_y)**2 + (j - center_x)**2)
    return (mat)
print ("Filter_Rise")
img = plt.imread("pypyimg.png")
m, n, z = img.shape
k = m * n
base = np.ones(k)
base.shape = (m, n)
new = Euclidist(base, m, n)
new = new / new.max()
new = new + 1
new = 1 / new

img[:, :, 2] *= new * 0.7
img[:, :, 3] *= new * 1
plt.imshow(img)
```

```python
#Filter_AMARO
def myfunct(nd):
    mat = np.linspace(0, 1, nd//2)
    mat.shape = (1, nd//2)
    return (mat)
def myfunct2(nd):
    mat = np.linspace(1, 0, nd//2)
    mat.shape = (1, nd//2)
    return (mat)
def myfunct3(nd):
    mat = np.linspace(0, 1, nd//2)
    mat.shape = (1, nd//2)
    mat = np.transpose(mat)
    return (mat)
def myfunct4(nd):
    mat = np.linspace(1, 0, nd//2)
    mat.shape = (1, nd//2)
    mat = np.transpose(mat)
    return (mat)
print ("Filter_Amaro")
img = plt.imread("pypyimg.png")
m, n, z = img.shape
base = myfunct(n)
base2 = myfunct2(n)
base3 = myfunct3(m)
base4 = myfunct4(m)
base = np.concatenate((base, base2), axis = 1)
for i in range(0, m - 1):
    n1 = myfunct(n)
    n2 = myfunct2(n)
    new = np.concatenate((n1, n2), axis = 1)
    base = np.concatenate((base, new))
base2 = np.concatenate((base3, base4), axis = 0)
for i in range(0, n - 1):
    n1 = myfunct3(m)
    n2 = myfunct4(m)
    new = np.concatenate((n1, n2), axis = 0)
    base2 = np.concatenate((base2, new), axis = 1)
base = base + base2
base = base / base.max()

for i in range(0, 3):
    img[:, :, i] *= base

plt.imshow(img)
```
